<b>Exercise 7:</b>
* Build the cross table between the features using the method <i>crosstab</i>.
* Visualize the distribution for the feature 'Churn', depending on the value of the features 'International plan', 'Voice mail plan', and 'Customer service calls'.
<b>Hint:</b> Use commands plt.sublot and sns.subplots.

Your plot should look something like this:

What do you see? What conclusions can be drawn? What feature (intuitively) can be more important for Churn prediction?

```python
rcparams = {'axes.titlesize':8, 'axes.labelsize':9, 'xtick.labelsize':6, 'ytick.labelsize':6, 'legend.fontsize':9,'figure.figsize':(10,5),'figure.dpi':100}
sns.set(rc = rcparams)
sns.set_style('white')

colors = ['#1f77b4', '#ff7f0e']
ct1 = pd.crosstab(df['Churn'],df['Voice mail plan'])
ct2 = pd.crosstab(df['Churn'],df['International plan'])
ct3 = pd.crosstab(df['Churn'],df['Customer service calls'])
fig, axs = plt.subplots(1,3)


ct1.plot(kind='bar', color=colors, ax=axs[0],legend = True)
axs[0].legend(title = 'Voice mail plan',loc='upper right')
axs[0].set_ylim(0,2800)

ct2.plot(kind='bar', color=colors, ax=axs[1],legend = True)
axs[1].legend(loc='upper right',title = 'International plan')
axs[1].set_ylim(0,2800)

ct3.plot(kind='bar', ax=axs[2])
axs[2].legend(loc='upper right',title = 'Customer service calls')
#axs[2].set_ylim(0,1200)

fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize = (12,4))
ct1.plot.bar(ax = axes[0])
ct2.plot.bar(ax = axes[1])
ct3.plot.bar(ax = axes[2])
```

### Observations

- Churn is more in the people who are subscribed to international plan than voice mail plan.
- If the customer has more than 3 csr calls they are more likely to Churn
